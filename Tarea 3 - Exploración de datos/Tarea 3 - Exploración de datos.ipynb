# Graficos Especificos a cada variable o grupos de variables

## X1

```python
plt.hist(df["X1"], bins=8, color='blue', alpha=0.7)

# Customize the plot
plt.title('Histograma de Creditos asignados')
plt.xlabel("Cantidad del Credito asignado (NTD)")
plt.ylabel('Frecuencia')
```

## X2

```python
df_default=filtered_df = df[df["Y"].isin([1])]

df_no_default=filtered_df = df[df["Y"].isin([0])]
```

```python
gender_counts = df['X2'].value_counts()
ax = gender_counts.plot(kind='bar', color=['blue', 'pink'])
for i, count in enumerate(gender_counts):
    ax.text(i, count, str(count), ha='center', va='bottom')
plt.xlabel('Genero')
plt.ylabel('Conteo')
plt.title('Distribucion de datos entre hombres y mujeres')
plt.xticks(ticks=[0, 1], labels=['Hombres', 'Mujeres'])
plt.show()
```

```python
gender_counts = df_default['X2'].value_counts()
ax = gender_counts.plot(kind='bar', color=['blue', 'pink'])
for i, count in enumerate(gender_counts):
    ax.text(i, count, str(count), ha='center', va='bottom')
plt.xlabel('Genero')
plt.ylabel('Conteo')
plt.title('Distribucion de datos entre hombres y mujeres con creditos que hacen default')
plt.xticks(ticks=[0, 1], labels=['Hombres', 'Mujeres'])
plt.show()
```

```python
gender_counts = df_no_default['X2'].value_counts()
ax = gender_counts.plot(kind='bar', color=['blue', 'pink'])
for i, count in enumerate(gender_counts):
    ax.text(i, count, str(count), ha='center', va='bottom')
plt.xlabel('Genero')
plt.ylabel('Conteo')
plt.title('Distribucion de datos entre hombres y mujeres con creditos que no hacen default')
plt.xticks(ticks=[0, 1], labels=['Hombres', 'Mujeres'])
plt.show()
```

```python
women_default_count = df_default[df_default["X2"] == 2]["X2"].count()
women_no_default_count = df_no_default[df_no_default["X2"] == 2]["X2"].count()

# Create a pie chart
plt.pie([women_default_count, women_no_default_count], labels=['Default', 'No Default'], autopct='%1.1f%%', colors=['lightcoral', 'lightskyblue'])

# Add a title
plt.title('Distribution of Women')

# Show the plot
plt.show()
```
