```python
import pandas as pd
import seaborn as sns
import numpy as np
from sklearn import metrics
import matplotlib.pyplot as plt

df = pd.read_csv('DataFrameFinal.csv')
pd.set_option('display.max_columns', None)
```

```python
corr = df.drop(columns=["X2","X3","X4","X5","X6","X7","X8","X9","X10","X11","X24","X25","X26","X27","X28","X29","Y"]).corr()
sns.heatmap(corr, cmap="Blues", annot=True)
```

```python
gender_counts = df['X2'].value_counts()
ax = gender_counts.plot(kind='bar', color=['blue', 'pink'])
for i, count in enumerate(gender_counts):
    ax.text(i, count, str(count), ha='center', va='bottom')
plt.xlabel('Genero')
plt.ylabel('Conteo')
plt.title('Distribucion de datos entre hombres y mujeres')
plt.xticks(ticks=[0, 1], labels=['Hombres', 'Mujeres'])
plt.show()
```
